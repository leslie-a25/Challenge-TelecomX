**Extraccion y carga de datos a utilizar**

```python
import pandas as pd
import requests
import seaborn as sns
import matplotlib.pyplot as plt

# URL de la API (versión raw para que Python lea el contenido directamente)
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Extraer los datos
response = requests.get(url)
data_json = response.json()

# Convertir a DataFrame de Pandas
df_telecom = pd.DataFrame(data_json)

# Ver las primeras filas y estructura
print(df_telecom.head())
print(df_telecom.info())
```

**Transformacion y limpieza**

```python
# Eliminar filas repetidas
df_telecom.drop_duplicates(inplace=True)

# Corregir la columna de dinero (TotalCharges)
df_telecom['TotalCharges'] = pd.to_numeric(df_telecom['TotalCharges'], errors='coerce').fillna(0)

# Limpiar nulos generales
df_telecom.dropna(inplace=True)
```

**gráficas**

```python
#Gráfico estratégico: Relación Contrato vs Churn
plt.figure(figsize=(10,5))
sns.countplot(data=df_telecom, x='Contract', hue='Churn')
plt.title('Evasión por Tipo de Contrato')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df_telecom, x='Contract', hue='Churn', palette='viridis')
plt.title('Evasión por Tipo de Contrato')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df_telecom[df_telecom['Churn'] == 'Yes']['MonthlyCharges'], label='Se fueron', shade=True)
sns.kdeplot(data=df_telecom[df_telecom['Churn'] == 'No']['MonthlyCharges'], label='Se quedaron', shade=True)
plt.title('Distribución de Cargos Mensuales')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(7, 7))
df_telecom['InternetService'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribución de Servicios de Internet')
plt.ylabel('')
plt.show()
```
